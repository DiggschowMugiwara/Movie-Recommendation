```python
movies = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DATA/movies.csv")
```

```python
#import re is for functioning shit
#re.sub() first parameter is kinda condn that should be replaced , and second is what it shld be replaced
#^ is like not shit , that means anything other than a-z , A-Z,0-9 , , sholud be replaced by (nothing)or erased yo!
import re

def clean_title(movies):
  return re.sub("[^a-zA-Z0-9 ]","",movies)
```

```python
#we srtored that replaced shit into the original yo!
movies["cleaned_title"] = movies["title"].apply(clean_title)
```

```python
#creating a searchbar kinda thing yo!
#Vectorizing is kinda converting words in numbers kinda shit , where we can match the numbers to check wheter they are similar words or naahh yo!
from sklearn.feature_extraction.text import TfidfVectorizer
Vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf = Vectorizer.fit_transform(movies["genres"])
tfidf
```

```python
#cosine similarity checks how similar are 2 shits yo!
#here we take a movie name , clean that shit , vectoize it , pass into cosine similarity , display similar shit
from sklearn.metrics.pairwise import cosine_similarity
title = "Adventure Fantasy"
# def srch(title):
  # title = clean_title(title)
```

```python
  Vect_Title = Vectorizer.transform([title])
  # Vect_Title
  similarity=cosine_similarity(Vect_Title,tfidf)
  similarity
  # #here incdices are of 2 d array and flatten does it to 1d array
  indices = np.argsort(similarity).flatten()
  indices
  new = indices[-100:]
  new
```

```python
xmovies =  movies.iloc[new][::-1]
```
